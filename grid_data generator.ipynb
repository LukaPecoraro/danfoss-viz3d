```python
grid_data = {}

grid = {
      "width": 500,
      "height": 500,
      "depth": 500
    }

grid_data["grid"] = grid
```

```python
boxes = []

width = 5
depth = 5

# populate bottom layer with 
for i in range(width):
    for j in range(depth):
        box = {"width": 100, "height": 100, "depth": 100, "position_y": 0}
        box["position_x"] = i * 100
        box["position_z"] = j * 100

        boxes.append(box)

# add the second layer
for i in range(width):
    for j in range(depth):
        box = {"width": 100, "height": 100, "depth": 100, "position_y": 110}
        box["position_x"] = i * 100
        box["position_z"] = j * 100

        boxes.append(box)

# add the third layer
for i in range(width-2):
    for j in range(depth-2):
        box = {"width": 100, "height": 100, "depth": 100, "position_y": 220}
        box["position_x"] = i * 100
        box["position_z"] = j * 100

        boxes.append(box)

# add the separators
separator = {"width": width * 100, "height": 10, "depth": depth * 100, "position_y": 100, "position_x" : 0, "position_z" : 0}
boxes.append(separator)
separator = {"width": width * 100, "height": 10, "depth": depth * 100, "position_y": 210, "position_x" : 0, "position_z" : 0}
boxes.append(separator)
grid_data["boxes"] = boxes
```
